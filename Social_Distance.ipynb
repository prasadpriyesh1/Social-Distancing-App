```python
import cv2
import math
```

```python
#some observations made by manually running some experiment on my laptop webcam
#all the dimensions are in inch
#actual width of the object
WIDTH = 57
# Perpendicular distance from the camera
DISTANCE = 38
#pixels covered by the object in the image
PIXELS = 480
# focus of the camera
# this value will be used in future calculations
FOCUS = PIXELS * DISTANCE / WIDTH
FOCUS
```

```python
#taking a general face dimensions
Face = (9,8)
faces = []
cap = cv2.VideoCapture(0)

```

```python
face_model = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
while True:
    status , image = cap.read()
    faces =[]
    face_cor = face_model.detectMultiScale(image)
    
    for face in face_cor:
        y = round((Face[0]*FOCUS/(face[2])),5)
        if face[0]   < 240:
            P = 480 - 2*(face[0] )
            P_ = y * P /38
            x = round((P_ * y / (FOCUS * 2)),5)
            x = -x
        elif face[0]   > 240:
            P = 2*(face[0] ) - 480
            P_ = y * P /38
            x = round((P_ *y / (FOCUS * 2)),5)
        else:
            x = 0
        faces.append((x,y))
    print(faces)
    warning = []
    for face in faces:
        warning.append([])
    if len(faces)>1:
        for a in range(len(faces)):
            for b in range(a + 1,len(faces)):
                dist = (faces[a][0] - faces[b][0])**2 + ( faces[a][1] - faces[b][1])**2
                dist = round(math.sqrt(dist),5)
                if dist < 40:
                    warning[a].append(b)
                    warning[b].append(a)
                    
    print(warning)
    str1 = ''
    n = 0
    if len(warning)>0:
        for war in range(len(warning)):
            
            str1 =str1 + '\n'
            str1 += 'Person '
            war3 = str(n)
            str1 += war3
            str1 += ': '
            for war2 in warning[war]:
                str1 += str(war2)
                str1 += ' , '
            n = n+1
    y0, dy = 50, 20
    for i, line in enumerate(str1.split('\n')):
        y1 = y0 + i*dy
        y1 = int(y1)
        cv2.putText(image, line , (0,y1) ,cv2.FONT_HERSHEY_SIMPLEX,0.5,[0,0,255],3)
    
    cv2.imshow('hello' , image)
    if cv2.waitKey(10) == 13:
        break
cv2.destroyAllWindows()
cap.release()
```
